```python
from pyspark.sql import SparkSession

spark = SparkSession.builder\
                    .appName('DPP-demo')\
                    .getOrCreate()
```

```python
dates_data = [
    ("2023-01-01", 2023),
    ("2023-01-02", 2023),
    ("2023-02-01", 2023),
    ("2022-12-01", 2022)
]

dates_df = spark.createDataFrame(dates_data, ["date", "year"])
```

```python
spark.conf.set("spark.sql.optimizer.dynamicPartitionPruning.enabled", "true")
```

```python
sales = spark.read.parquet("sales_data")

filtered_dates = dates_df.filter("year = 2023")

result = sales.join(filtered_dates, "date")
```
