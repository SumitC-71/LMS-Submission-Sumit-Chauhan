```python
'''
use cases:
specific data extract from log files

validate valid emails 

validate phone no format
(filter phone no of only india or new york)

find all area codes

supported by different languages

RE are search pattern used to search a pattern in large text

text and RE both has their own cursor that moves if both are equal
'''

import re
```

```python
NameAge = '''
Alice is 22 and John is 24
Bob is 23
'''
print(NameAge)
ages = re.findall(r'\d{1,3}',NameAge)
names = re.findall(r'[A-Z][a-z]*',NameAge)

print(ages)
print(names)

x = 0
ageDict={}
for name in names:
    ageDict[name] = ages[x]
    x+=1

print(ageDict)
```

```python
flag = re.search('secret','There is some top secret')
print(flag)
print(type(flag))

if flag:
    print('There is secret')
else:
    print('There is no secret')
```

```python
allSecrets = re.findall('secret', 'There is secret1, secret2 and secret2')
# findall returns list

for secret in allSecrets:
    print(secret)
```

```python
str = 'There is secret1, secret2 and secret2'
for i in re.finditer('secret',str):   # you iterate through iterators
    loctup = i.span() # returns tuple: (starting index, ending index)
    print(loctup)
```

```python
str = 'sat, hat, pat, mat, ald'

allAt = re.findall('[shpm]at',str)

print(allAt)
```

```python
str = 'sat, hat, pat, mat, ald'

allAt = re.findall('[h-z]at',str)

print(allAt)
```

```python
str = 'sat, hat, pat, mat, ald'

allAt = re.findall('[^h-m]at',str)   # ^ symbol means negation that is everything apart from (h to m) 

print(allAt)   
```

```python
# find and replace
food = 'hat rat mat pat'
regex = re.compile('[r]at')  # creates re.Pattern object
print(type(regex))

food = regex.sub('food',food)

print(food)
```

```python
randstr = 'this is \\abc'

print(re.search(r'\\abc',randstr))

# for matching (double backslash)
# this is called backslash problem
# we use raw search for it
```

```python
# replacing whitespace chars
randstr = '''
hello
how are you
?
'''
print(randstr)
regex = re.compile("\n")
randstr = regex.sub(' ',randstr)
print(randstr)
```
