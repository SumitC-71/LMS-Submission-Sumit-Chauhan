```python
datetime.datetime(2019,6,7,23,54,11)
print(datetime.datetime(2019,6,7,23,54,11))
```

```python
print(datetime.datetime.today())
```

```python
print(datetime.datetime.now())  # same as today()
```

```python
v = datetime.datetime.now()
print(v.year)
print(v.month)
print(v.date()) # 2026-03-10
print(v.time()) # 14:34:46.615644
```

```python
a = datetime.timedelta(days=30,hours=2)  # similar to interval in sql
b = datetime.timedelta(days=29,hours=1)

print(a - b)
```

```python
today = datetime.datetime.today() + datetime.timedelta(days=1)
print(today)
```

```python
print(today.strftime('%Y/%m/%d'))
```

```python
from datetime import datetime

d1 = datetime(2026,3,10)
d2 = datetime(2026,3,1)

diff = d1 - d2

print(diff)
print(type(diff)) # diff is of type datetime.timedelta
print(diff.days)
print(diff.seconds)

print(diff.total_seconds()) # total seconds in difference

```

```python
s = '15 August 2016'
str_local = datetime.strptime(s,'%d %B %Y')  # converts string to struct_date 
print(str_local)

dt = str_local.date()  # you can extract date or time or specific field like year from struct_date
print(dt)

print(type(dt))
dt = dt.strftime('%Y/%m/%d')  # converts date to string
print(dt)
print(type(dt))
```
