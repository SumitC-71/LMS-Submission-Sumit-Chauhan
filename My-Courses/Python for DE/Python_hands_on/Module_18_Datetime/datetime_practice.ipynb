```python
import datetime
```

```python
datetime.datetime(2019,6,7,23,54,11)
print(datetime.datetime(2019,6,7,23,54,11))
```

```python
print(datetime.datetime.today())
```

```python
print(datetime.datetime.now())  # same as today()
```

```python
v = datetime.datetime.now()
print(v.year)
print(v.month)
print(v.date()) # 2026-03-10
print(v.time()) # 14:34:46.615644
```

```python
a = datetime.timedelta(days=30,hours=2)  # similar to interval in sql
b = datetime.timedelta(days=29,hours=1)

print(a - b)
```

```python
today = datetime.datetime.today() + datetime.timedelta(days=1)
print(today)
```
