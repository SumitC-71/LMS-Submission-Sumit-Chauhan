```python
'''
naiive 
'''
import datetime

d = datetime.date(year=2016, month=7, day=10) # zero padded automatically
print(d)

td = datetime.date.today() 
print(type(td))

print(td.weekday()) # monday 0 sunday 6
print(td.isoweekday()) # monday 1 sunday 7

```

```python
tdelta = datetime.timedelta(days=7)
print(type(td))
print(type(tdelta))

print(type(td + tdelta))   # date + timedelta


#  date1 = date + timedelta
#  timedelta = date1 + date1

# datetime = datetime + timedelta 
```

```python
import  pytz

dt = datetime.datetime(2026, 3, 13, 10, 39, 40, tzinfo=pytz.UTC)
print(dt)

dt_now = datetime.datetime.now(tz=pytz.UTC) # timezone aware
print(dt)

dt_now_pacific = dt_now.astimezone(pytz.timezone('Asia/Calcutta'))  # dt_now must be timezone aware otherwise you will get error     

print(dt_now_pacific)
```

```python
# listing all timezones
for tz in pytz.all_timezones:
    print(tz)
```

```python
naive_dt = datetime.datetime.now()
print(naive_dt)
tz_aware_dt = naive_dt.astimezone(pytz.timezone('Asia/Colombo'))
print(tz_aware_dt) 
```

```python
# localize
naive_dt = datetime.datetime.now()

print(naive_dt)
tz_factory = pytz.timezone('Asia/Bangkok')
naive_dt = tz_factory.localize(naive_dt)

print(naive_dt)
```
