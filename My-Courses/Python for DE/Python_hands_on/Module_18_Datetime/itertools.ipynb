```python
from itertools import count

for i in count(10, 2):
    print(i)
    if i > 20:
        break
```

```python
from itertools import cycle

c = cycle(['A','B','C'])

for i in range(7):
    print(next(c))
```

```python
from itertools import repeat

list(repeat("hello", 3))

```

```python
from itertools import chain

list(chain([1,2,3], [4,5], [6]))

lst = [1,2,3]
lst2 = [4,5]
lst3 = [6]

# or
print(lst + lst2 + lst3)
```

```python
from itertools import islice

nums = range(100)

list(islice(nums, 5, 10))  # format: iter-seq, start, stop (excluded)
```

```python
from itertools import accumulate

# list(accumulate([1,2,3,4]))
list(accumulate((1,2,3,4)))
```

```python
import operator
list(accumulate([1,2,3,4], operator.mul))
```

```python
from itertools import groupby

data = [1,1,1,2,2,3,3,3]

print(groupby(data))

for key, group in groupby(data):
    print(key, tuple(group))
```

```python
from itertools import product

list(product([1,2], ['A','B']))
```

```python
from itertools import permutations

list(permutations([1,2,3]))
```

```python
from itertools import combinations

list(combinations([1,2,3,4],2))   # 4C2 combinations
```

```python
from itertools import tee

a, b = tee([1,2,3,4])

print(list(a))
print(list(b))
```

```python
from itertools import zip_longest
a = [1,2,3,4,5]
b = ['a','b','c']

for i,j in zip_longest(a,b,fillvalue='X'):
    print(i,j)
```

### merging 3 files using chain

```python
from itertools import chain

with open('test1.txt') as r1, \
     open('test2.txt') as r2, \
     open('test3.txt') as r3, \
     open('merged_test.txt','w') as w:

    w.writelines(chain(r1, r2, r3))
```
