# Merge tables

````python
new_data = [
    (1, "Aakash", "IT", 120000),
    (6, "Simran", "HR", 45000)
]

new_df = spark.createDataFrame(new_data, columns)

new_df.show()
````

````python
new_df.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("new_employees")
````

````python
%sql

MERGE INTO employees AS target
USING new_employees AS source
ON target.id = source.id

WHEN MATCHED THEN
UPDATE SET *

WHEN NOT MATCHED THEN
INSERT *
````

# Update and Delete using spark code

````python
from delta.tables import DeltaTable

delta_table = DeltaTable.forName(spark, "employees")

'''
forPath() | External/path-based tables            | /tmp/..., /FileStore/...
forName() | Managed tables (Unity Catalog / Hive) | "employees"
'''

delta_table.update(
    condition="id = 1",
    set={"salary": "30000"}
)

delta_table.delete("id = 2")
````

````python
from delta.tables import DeltaTable

delta_table = DeltaTable.forName(spark, 'employees')

data = [(1, 'John', 'Sales', 40000), (2, 'Sally', 'Engineering', 50000)]

source_df = spark.createDataFrame(data,["id", "name", "department", "salary"])

delta_table.toDF().show()
````

````python
delta_table.alias('t').merge(
    source_df.alias('s'),
    't.id = s.id'
).whenNotMatchedInsertAll()\
    .whenMatchedUpdateAll()\
    .execute()
````

````python
delta_table.toDF().show()
````

VACUUM → remove old files
